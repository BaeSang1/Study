```python
def hello(name: str) -> None:
  if not isinstance(name,str):
    raise TypeError("Name Must be a string!")
  print(f"안녕! {name}")
```

```python
def add(*args):
  print(args,type(args))
add(1,2,3)
```

```python
def add(*args):
  addition = 0
  for number in args:
    addition += number
  return addition

print(add(1,2,3))
print(add(1,2))
print(add(1,2,3,4))
```

```python
# 머신러닝 클래스, 하이퍼 파라미터 튜닝 시, 입력값이 딕셔너리를 기본값으로 받음
def temp(**kwargs):
  for key,value in kwargs.items():
    print(key,value)

temp(name="SangIl",age=30,city="Suwon")
```

# shift()
- 시계열 데이터의 값 비교를 위해 한 칸씩 밀며 새로운 열 생성


```python
temp_df = sales[['datesold','price']].copy()
temp_df['shifted_v1'] = temp_df['price'].shift(1,fill_value=0).astype(int)
temp_df['shifted_v2'] = temp_df['price'].shift(2,fill_value=0).astype(int)

temp_df['차이값'] = temp_df['price'] - temp_df['shifted_v1']
temp_df.head()
```

```python
# 0000년 00월 00일
temp_df['한글날짜'] = temp_df['datesold'].dt.strftime('%Y년 %m월 %d일')
temp_df.head()
```
